```python
import numpy as np
import matplotlib.pyplot as plt
```

# Simulating Montecarlo

```python
#Generating a Montecarlo simulation of a fake stock price
#Number of iterationrs
number_of_periods=10
number_of_simulations=10000
#This is the Initial State
mu, sigma = 0, 0.5
#Run the simulation n times
for i in range(number_of_simulations):
    Price_list=[1]
    Periods=list(range(0, number_of_periods+1))
    for i in range(number_of_periods):
        s = np.random.normal(mu, sigma, 1) # You change the distribution , or you use a machine learning model as a data generation process
        change=s[0]
        next_price=Price_list[-1]+change
        Price_list.append(next_price)
        
    plt.plot(Periods,Price_list)
#Graph characteristics
plt.title('Fake Data Graph')
plt.xlabel('Periods')
plt.ylabel('Price $')
plt.show()
```

# Simulating a biological system

```python
#Initial Number of Pray
x_t=[50]
#Initial Number of Predator
y_t=[1]
#Parameters
Alpha=3
Beta=0.5
Gamma=2
Delta=0.5

#Number of iterationrs
number_of_periods=3
Periods=list(range(0, number_of_periods+1))
#This is the Initial State
#Run the simulation n times
for i in range(number_of_periods):
    Change_X=Alpha*x_t[-1]-Beta*x_t[-1]*y_t[-1]
    Change_Y=Gamma*y_t[-1]-Delta*x_t[-1]*y_t[-1]
    next_x_t=x_t[-1]+Change_X
    if next_x_t<0:
        next_x_t=0
    else:
        pass
    next_y_t=y_t[-1]+Change_Y
    if next_y_t<0:
        next_y_t=0
    else:
        pass
    x_t.append(next_x_t)
    y_t.append(next_y_t)
          
plt.plot(Periods,x_t)
plt.plot(Periods,y_t)
#Graph characteristics
plt.title('Population Dynamics')
plt.legend(['Pray', 'Predators'])
plt.xlabel('Periods')
plt.ylabel('Number of Individuals')
plt.show()

x_t
```
